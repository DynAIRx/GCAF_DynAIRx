```python
import pandas as pd
import os
```

```python
input_files_folder="Codelists_Completed_after_Review"
filepath_merged_output="merged_codelists_with_snomed.csv"
file_list = os.listdir(input_files_folder)
#file_list = ['grouped_database_with_snomed_Abdominal Aortic Aneurysm.csv', 'grouped_database_with_snomed_Back pain.csv']
file_list
```

```python
df = pd.DataFrame()
for filename in file_list:
    filepath=input_files_folder+"/"+filename
    print("Processing file {}".format(filename))
    data = pd.read_csv(filepath, dtype={'SnomedCTConceptId': str})
    df = pd.concat([df, data], axis=0)

df.to_csv(filepath_merged_output, index=False)
print("Merging Completed!!! Check file {}".format(filepath_merged_output))
```
