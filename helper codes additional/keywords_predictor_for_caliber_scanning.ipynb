```python
import os

codelist_files_folder="Caliber/Caliber_Codelist_Files_with_SNOMEDs_Added"
#"Caliber\hdr-caliber-phenotype-library-master\_data\codelists"
codelist_files = os.listdir(codelist_files_folder)
codelist_files
```

```python
codelist_names=[]
for filename in codelist_files:
    filename_parts = filename.split("_")
    len_parts=len(filename_parts)
    new_filename=""
    for i in range(len_parts-2):
        new_filename= new_filename+" "+filename_parts[i]
    codelist_names.append(new_filename)
print(codelist_names)
```

```python
# f = open("caliber_condition_names.csv", "w")
# f.write(codelist_names)
# f.close()
# import csv

# with open('caliber_condition_names.csv', 'w', newline='') as file:
#     writer = csv.writer(file)
#     writer.writerows(codelist_names)

import csv
rows = zip(codelist_names)
with open('temp_caliber_condition_names.csv', "w") as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
