```python
codelist_files = os.listdir(codelist_files_folder)
codelist_files
```

```python
df_database_efi = pd.read_csv(filepath_database_efi, dtype={'SNOMEDCT_CONCEPTID': str})  #pd.read_excel(filepath_database_efi, sheet_name=0)
df_database_efi
```

```python
list_dest_snomeds = df_database_efi['SNOMEDCT_CONCEPTID'].tolist()
print(list_dest_snomeds)
print("Number of SNOMEDs in efi file: ", len(list_dest_snomeds))
```

```python
for filename in codelist_files:
    our_codelist_filepath=codelist_files_folder+"/"+filename
    df_source_file = pd.read_csv(our_codelist_filepath, dtype={'SnomedCTConceptId': str}) 
    list_source_snomeds = df_source_file['SnomedCTConceptId'].tolist()
    print("------For "+filename)
    additional_snomed_codes=[]
    for snomed in list_source_snomeds:
        #print(snomed)
        if snomed in list_dest_snomeds:
            #print("Present ", snomed)
            snomed_indices=df_database_efi.index[df_database_efi['SNOMEDCT_CONCEPTID'] == snomed].tolist()
            #print(snomed_indices)

            code_description_absent=pd.isnull(df_database_efi.loc[snomed_indices[0], 'Codedescription'])# check if code description value is null at index--
            #print(snomed, df_database_efi.loc[snomed_indices[0], 'Codedescription'])

            if code_description_absent:
                print("Need Vaidation")
        else:
            print("Absent ", snomed)
            additional_snomed_codes.append(snomed)



    if len(additional_snomed_codes)==0:
        print("Congratulations all snomed codes were present for this file: ", our_codelist_filepath)
        #df_shrinked = df_shrinked.drop(df_shrinked[df_shrinked['SnomedCTConceptId'] == snomed].index)
    else:   #----this code shrink files for identified snomed's---
        shrinked_filepath=filepath_output+"/Shrinked "+filename
        df_shrinked = pd.read_csv(our_codelist_filepath, dtype={'SnomedCTConceptId': str}) 
        list_source_snomeds = df_shrinked['SnomedCTConceptId'].tolist()
        print(additional_snomed_codes)
        for snomed in list_source_snomeds:
            if snomed in additional_snomed_codes:
                continue
            else:
                 df_shrinked = df_shrinked.drop(df_shrinked[df_shrinked['SnomedCTConceptId'] == snomed].index)
        df_shrinked.to_csv(shrinked_filepath, index=False)  
```
