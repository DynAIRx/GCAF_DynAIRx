```python
# Read Excel file
df_comments = pd.read_excel('Clincians_Intervention_Split_Keep_Group.xlsx')
df_comments
```

```python
# Function to separate column1 values based on keywords in column2
def separate_by_keywords(row):
    # Categorize Column1 based on Column2
    if pd.isna(row['Comments_eFI']):
        group1 = 'Other'
    elif 'compare'.lower() in str(row['Comments_eFI']).lower():
        group1 = 'Combine'
    elif 'keep'.lower() or 'include'.lower() in str(row['Comments_eFI']).lower():
        group1 = 'Keep'
    elif 'split'.lower() in str(row['Comments_eFI']).lower():
        group1 = 'Split'
    elif 'group'.lower() in str(row['Comments_eFI']).lower():
        group1 = 'Combine'
    else:
        group1 = 'Other'
    
    group2 = 'Other'  # Default value for group2 in case no condition matches
    # Categorize Column3 based on Column4
    if pd.isna(row['Comments_LW']):
        group2 = 'Other'
    elif 'compare'.lower() in str(row['Comments_eFI']).lower():
        group1 = 'Combine'
    elif 'keep'.lower() in str(row['Comments_LW']).lower():
        group2 = 'Keep'
    elif 'split'.lower() in str(row['Comments_LW']).lower():
        group2 = 'Split'
    elif 'group'.lower() in str(row['Comments_LW']).lower():
        group2 = 'Combine'
    else:
        group2 = 'Other'
    
    return row['eFI2 Deficits'], row['Comments_eFI'], group1, row['Liverpool Diseases'], row['Comments_LW'], group2
```

```python
# Apply the function to create new columns for Group1 (Column1 based on Column2) and Group2 (Column3 based on Column4)
df_comments[['eFI2 Deficits', 'Comments_eFI', 'Instructions_for_eFI', 'Liverpool Diseases', 'Comments_LW', 'Instructions_for_LW']] = df_comments.apply(separate_by_keywords, axis=1, result_type="expand")


# # Define the new column order
new_order = ['eFI2 Deficits', 'Comments_eFI', 'Instructions_for_eFI', 'Liverpool Diseases', 'Comments_LW', 'Instructions_for_LW']
# # Reorder the DataFrame columns
df_comments_with_automatic_instructions = df_comments[new_order]

# View the updated DataFrame
df_comments_with_automatic_instructions
```

```python
# save the result to a new Excel file
df_comments_with_automatic_instructions.to_excel('Output/GCAF_Generated_Distribution.xlsx', index=False)
```
